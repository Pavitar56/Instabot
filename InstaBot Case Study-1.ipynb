### Extracting usernames of top 500 followers of foodtalkindia by scrolling through them 

```python
while True:
    try:
        foodtalkindia=[]
        count=0
        while True:
            
            elements=driver.find_elements_by_xpath('//a[contains(@class, "notranslate")]')
            
            if len(elements)<500:
                waiter=WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//ul[contains(@class, "jSC57")]')))
                driver.execute_script('arguments[0].scrollIntoView(0, 100);', driver.find_element_by_xpath('//ul[contains(@class, "jSC57")]'))
                time.sleep(0.8)
            
            i=elements[count]
            foodtalkindia.append(i.get_attribute('innerHTML'))
            
            if len(foodtalkindia)>=500:
                break
            
            count+=1
        
        break
    
    except StaleElementReferenceException:
        continue

driver.back()
time.sleep(3)
foodtalkindia
```

### Now performing task 6.2 by first getting all users who I m following but they don't follow me and then comparing them with followers of foodtalkindia to get final result..Followers of foodtalkindia are already stored in list got in task 6.1 above

```python
my_following=set(my_following)
my_followers=set(my_followers)
foodtalkindia=set(foodtalkindia)  #followers of foodtalkindia

people_i_follow_but_they_dont_follow_me=my_following.difference(my_followers)
people_i_follow_but_they_dont_follow_me_but_follow_foodtalkindia=people_i_follow_but_they_dont_follow_me.intersection(foodtalkindia)
if len(people_i_follow_but_they_dont_follow_me_but_follow_foodtalkindia)==0:
    print("No such people")
else:
    for i in people_i_follow_but_they_dont_follow_me_but_follow_foodtalkindia:
        print(i)
```
